```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_graph(G, title):
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, edge_color='k', font_size=15, font_weight='bold')
    plt.title(title)
    plt.show()
```

### Simple Graph

```python
simple_graph = nx.Graph()
simple_graph.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
plot_graph(simple_graph, "Simple Graph")
```

A simple graph is an undirected graph that does not contain any loops (edges connected at both ends to the same vertex) or multiple edges between the same pair of vertices.

### Directed Graph

```python
directed_graph = nx.DiGraph()
directed_graph.add_edges_from([(1, 2), (2, 3), (3, 1), (3, 4)])
plot_graph(directed_graph, "Directed Graph")
```

A directed graph, or digraph, consists of vertices connected by directed edges, where the edges have a direction associated with them.

### Complete Graph

```python
complete_graph = nx.complete_graph(5)
plot_graph(complete_graph, "Complete Graph")
```

A complete graph is a simple undirected graph in which every pair of distinct vertices is connected by a unique edge.

### Bipartite Graph

```python
G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4], bipartite=0)
G.add_nodes_from(['A', 'B', 'C'], bipartite=1) 

G.add_edges_from([(1, 'A'), (2, 'A'), (3, 'B'), (4, 'C'),(2, 'B'),(3, 'C')])
pos = nx.bipartite_layout(G, [1, 2, 3, 4])
nx.draw(G, pos=pos, with_labels=True)

plt.title("Bipartite Graph")
plt.show()
```

A bipartite graph is a type of graph where the set of nodes can be divided into two disjoint sets such that no two nodes within the same set are adjacent. In other words, it's a graph where every edge connects a node from one set to a node in the other set. Bipartite graphs are often used to represent relationships between two different types of entities.

### Subgraph Graph

```python
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (4, 6)])

subgraph_nodes = [1, 2, 3]
subgraph = G.subgraph(subgraph_nodes)

pos = nx.spring_layout(G)
nx.draw(G, pos=pos, with_labels=True, node_color='skyblue')
nx.draw(subgraph, pos=pos, with_labels=True, node_color='salmon')

plt.title("Subgraph Graph")
plt.show()
```

A subgraph is a portion of a larger graph that consists of a subset of nodes and edges from the original graph. It retains the structure of the original graph, but with fewer nodes and edges. Subgraphs are useful for analyzing specific components or relationships within a larger graph.

### Travelling in Graph

```python
import matplotlib.pyplot as plt
import networkx as nx

walk = nx.Graph()
walk.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)])
trail = nx.Graph()
trail.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 3)])
path = nx.Graph()
path.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 3)])
closed_walk = nx.Graph()
closed_walk.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
circuit = nx.Graph()
circuit.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])
simple_circuit = nx.Graph()
simple_circuit.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

walk_path = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)]
trail_path = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 3)]
path_path = [(1, 2), (2, 3), (3, 4), (4, 5)]
closed_walk_path = [(1, 2), (2, 3), (3, 4), (4, 1)]
circuit_path = [(1, 2), (2, 3), (3, 4), (4, 1)]
simple_circuit_path = [(1, 2), (2, 3), (3, 4), (4, 1)]


plt.figure(figsize=(15, 10))
plt.subplot(236)
nx.draw(walk, pos=nx.circular_layout(walk), with_labels=True)
nx.draw_networkx_edges(walk, pos=nx.circular_layout(walk), edgelist=walk_path, edge_color='red', width=2)
plt.title('Walk')

plt.subplot(235)
nx.draw(trail, pos=nx.planar_layout(trail), with_labels=True)
nx.draw_networkx_edges(trail, pos=nx.planar_layout(trail), edgelist=trail_path, edge_color='red', width=2)
plt.title('Trail')

plt.subplot(234)
nx.draw(path, pos=nx.spring_layout(path, seed=57), with_labels=True)
nx.draw_networkx_edges(path, pos=nx.spring_layout(path, seed=57), edgelist=path_path, edge_color='red', width=2)
plt.title('Path')

plt.subplot(233)
nx.draw(closed_walk, pos=nx.spring_layout(closed_walk, seed = 58), with_labels=True)
nx.draw_networkx_edges(closed_walk, pos=nx.spring_layout(closed_walk, seed = 58), edgelist=closed_walk_path, edge_color='red', width=2)
plt.title('Closed Walk')

plt.subplot(232)
nx.draw(circuit, pos=nx.spring_layout(circuit, seed = 59), with_labels=True)
nx.draw_networkx_edges(circuit, pos=nx.spring_layout(circuit, seed = 59), edgelist=circuit_path, edge_color='red', width=2)
plt.title('Circuit')

plt.subplot(231)
nx.draw(simple_circuit, pos=nx.spring_layout(simple_circuit, seed = 60), with_labels=True)
nx.draw_networkx_edges(simple_circuit, pos=nx.spring_layout(simple_circuit, seed = 60), edgelist=simple_circuit_path, edge_color='red', width=2)
plt.title('Simple Circuit')

plt.tight_layout()
plt.show()
```

"Travelling in a graph" typically refers to various concepts related to moving through a graph from one vertex to another. This can involve paths, walks, trails, circuits, or cycles, depending on the context and specific constraints. Here are some key concepts related to traveling in a graph:

Walk: A walk is a sequence of vertices and edges, where vertices and edges can be repeated.  
Trail: A trail is a walk in which no edge is repeated, though vertices can be repeated.  
Path: A path is a sequence of edges that connects a sequence of distinct vertices.  
Closed Walk: A closed walk is a walk from v to w that starts and ends in the same vertex.  
Circuit: A circuit is a trail that starts and ends at the same vertex.  

### The Handshake Theorem

The Handshake Theorem is a simple yet powerful tool in graph theory, highlighting the relationship between vertex degrees and edges in an undirected graph. It ensures the consistency of graph structures and aids in the analysis and understanding of various graph properties.  

<b>Degree of a Vertex:</b> The degree of a vertex is the number of edges incident to (connected to) that vertex.  
<b>Sum of Degrees:</b> When you sum the degrees of all vertices, each edge is counted twice (once for each endpoint of the edge). Thus, the total sum of the degrees is twice the number of edges.

### Connectedness of a Graph

Connectedness is a fundamental concept in graph theory that describes the extent to which the vertices of a graph are reachable from one another. Understanding whether a graph is connected, and identifying its connected components, are crucial steps in analyzing the structure and properties of the graph.

### Connected Components

A graph $H$ is connected component of a graph $G$ $\Longleftrightarrow$:

- $H$ is a subgraph of $G$.
- $H$ is connected.
- No connected subgraphs of $G$ has $H$ as a subgraphs and contains vertices or edges that are not in $H$.

### Euler's Circuit

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.cycle_graph(8)
euler_circuit = list(nx.eulerian_circuit(G))

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=1000, font_size=12, font_weight='bold')

plt.title(f"Has Eulerian Circuit: {euler_circuit}")
plt.show()

```

An Eulerian circuit is a path in a graph that traverses every edge exactly once and returns to the starting point. In simpler terms, it's a route through a graph that covers all the edges without repeating any, allowing you to start and end at the same node.

### Eulerian Trail

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph with an Eulerian trail
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (4, 5),(1, 5)])  # Example of an Eulerian trail

# Get an Eulerian trail
euler_trail = list(nx.eulerian_path(G))

# Plot the graph with the Eulerian trail
plt.figure(figsize=(6, 4))
pos = nx.circular_layout(G)  # Position nodes in a circle
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edges(G, pos, edgelist=euler_trail, edge_color='red', width=2)
plt.title("Eulerian Trail")
plt.show()

```

An Eulerian trail is a path in a graph that traverses every edge exactly once but may not necessarily start and end at the same vertex. In simpler terms, it's a route through a graph that covers all the edges without repeating any, allowing you to start at one vertex and end at another.

### Hamiltonian Circuit

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.cycle_graph(7)

# Add some extra edges to make it more complex
extra_edges = [(0, 3), (1, 4), (2, 5), (3, 6)]
G.add_edges_from(extra_edges)

# Find a Hamiltonian circuit
hamiltonian_circuit = list(G.nodes()) + [list(G.nodes())[0]]  # Hamiltonian circuit

# Plot the graph with the Hamiltonian circuit
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Position nodes with a spring layout
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edges(G, pos, edgelist=[(hamiltonian_circuit[i], hamiltonian_circuit[i+1]) for i in range(len(hamiltonian_circuit)-1)], edge_color='red', width=2)
plt.title("Hamiltonian Circuit")
plt.show()

```

A Hamiltonian circuit is a path in a graph that visits each node exactly once and returns to the starting node, forming a closed loop. Think of it like a tour that covers all the stops in a city without repeating any, and eventually brings you back to where you started.

### Travelling Salesman Problem Graph

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
positions = {
    1: (0, 0),
    2: (1, 1),
    3: (2, 0),
    4: (1, -1),
    5: (3, 1),
    6: (4, 0),
}

for node, pos in positions.items():
    G.add_node(node, pos=pos)

edges = [
    (1, 2, 1.4),
    (1, 3, 2),
    (1, 4, 1.4),
    (2, 3, 1.4),
    (2, 4, 2),
    (2, 5, 2.2),
    (3, 4, 1.4),
    (3, 6, 2.2),
    (4, 6, 3),
    (5, 6, 1.4),
]

for edge in edges:
    G.add_edge(edge[0], edge[1], weight=edge[2])

pos = nx.get_node_attributes(G, 'pos')
nx.draw_networkx_nodes(G, pos, node_size=700)

edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edges(G, pos, edgelist=edges, width=2)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

plt.title("Traveling Salesman Problem Graph")
plt.axis('off')
plt.show()
```

A Traveling Salesman Problem graph is a network of nodes connected by edges, where each edge has a weight representing the distance or cost of traveling between nodes. The goal is to find the shortest possible route that visits each node exactly once and returns to the starting point. The graph visually represents these nodes and weighted edges to help solve the problem.
